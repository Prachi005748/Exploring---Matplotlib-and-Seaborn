Matplotlib Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x **2
y
```

```python
plt.plot(x,y)
plt.show
plt.title("Hello World")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Subplots in matplotlibs

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x **2,x **2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

Object Oriented Way

```python
fig = plt.figure(figsize=(20,8),dpi = 100)
axes1 = fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2 = fig.add_axes([0.3,0.4,.3,.3])
axes2.plot(x,y)
```

1. Which team won the most matches?

2. Toss Decision Trends 

```python
sns.countplot(x = df['toss_decision'],palette = 'rainbow')
plt.title("Toss Decision Trends")
```

3. Toss Winner vs Match Winner

```python
count = df[df['toss_winner']== df ['match_winner']]['match_id'].count()
percentage = (count *100)/df.shape[0]
percentage
```

4. How do teams win?(Runs vs wickets)

```python
sns.countplot(x = df['won_by'],palette = 'rainbow')
plt.title("Won By")
```

1. Most "Player of the match " Award

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values,y = count.index,palette = 'mako')
plt.title("Top 10 players with man of thr match")
```

2. Two Top Scorers 

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

3. 10 Best Bowling Figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling') ['highest_wickets'].sum().sort_values(ascending = False).head(10)
top_bowlers.plot(kind = 'barh')
```
